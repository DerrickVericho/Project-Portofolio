```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://studentorg.ucsd.edu/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('table', class_ = 'table styled')
```

```python
table = soup.find_all('table')[0]
table
```

```python
link = table.find_all('tr')
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['Link'])
df
```

```python
for data in link:
    a_tags = data.find_all('a')
    for a_tag in a_tags:
        link_url = a_tag.get('href')
        length = len(df)
        df.loc[length] = url + link_url
```

```python
df
```

```python
df1 = pd.DataFrame(columns = ['Email'])
```

```python
for i in df['Link']:
    page1 = requests.get(i)
    soup = BeautifulSoup(page1.text, 'html')
    tag = soup.find_all('strong')[2]
    email = tag.text
    
    length = len(df1)
    df1.loc[length] = email
```

```python
df2 = pd.DataFrame(columns = ['Name of Organization'])
```

```python
for i in df['Link']:
    page1 = requests.get(i)
    soup = BeautifulSoup(page1.text, 'html')
    p = soup.find_all('article')[0]
    judul = p.find('h1').text
    
    length = len(df2)
    df2.loc[length] = judul
```

```python
df2
```

```python
df = pd.concat([df2, df1], axis = 1)
df.to_csv(r"C:\Users\ASUS\Downloads\Example Project.csv", index=False)
```
